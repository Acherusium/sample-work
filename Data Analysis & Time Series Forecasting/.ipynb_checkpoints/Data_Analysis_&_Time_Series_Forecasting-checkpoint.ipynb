# Turning Basic **Data** into **Information**

## Stores / City

###Grouping ***stores_by_city***

```python
# Group by 'city' and count the number of stores in each city
stores_by_city = stores.groupby('city')['store_nbr'].count().reset_index()
stores_by_city.rename(columns={'store_nbr': 'stores'}, inplace=True)
```

In this code block, data manipulation operation is performed on the 'stores' DataFrame. The objective is to group the data by the 'city' column and calculate the number of stores in each city.

### Display ***stores_by_city***

```python
# Create the bar chart
fig = px.bar(stores_by_city, x='stores', y='city',
             text_auto=True,
             labels={'city': 'City', 'stores': 'Number of Stores'})

# Calculate the mean number of stores
mean_stores = stores_by_city['stores'].mean()

# Add a line representing the mean number of stores
fig.add_shape(type='line',
              x0=mean_stores, y0=-1,
              x1=mean_stores, y1=len(stores_by_city),
              line=dict(color='red', width=2))

fig.update_traces(textfont_size=9, textangle=0, textposition="outside", cliponaxis=False)

fig.show()
```

##Stores / State & Type

###Grouping ***stores_by_state_and_type***

```python
# Group by 'state' and 'type' to get the count of store_nbr for each type in each state
stores_by_state_and_type = stores.groupby(['state', 'type'])['store_nbr'].count().reset_index()
```

This code block illustrates a more advanced data manipulation operation on the 'stores' DataFrame. The primary objective is to group the data by both the 'state' and 'type' columns and calculate the count of store numbers ('store_nbr') for each store type in each state.

This data manipulation operation allows for a comprehensive understanding of the distribution of store types across different states, providing valuable insights for business planning, resource allocation, and strategic decision-making.

```python
stores_by_state_and_type
```

###Display ***stores_by_state_and_type***

```python
fig = px.bar(stores_by_state_and_type, x="store_nbr", y="state",
             color="type", text_auto=True,
             title='Number of Stores per State & Type',
             labels={'state':'States', 'store_nbr': 'Number of Stores', 'type': 'Type'})
fig.update_traces(textfont_size=9, textangle=0, textposition="inside", cliponaxis=False)
fig.show()
```

####Percentage of Stores per Type

```python
fig = px.pie(stores_by_state_and_type, values='store_nbr', names='type',
             title='% of Stores per Type',)
fig.show()
```

#Merging **Datasets**

```python
# Merge transactions and stores datasets
merged_data = transactions.merge(stores, on='store_nbr')
```

The integration of transactional data with corresponding store details enables a holistic view of sales performance, allowing businesses to understand how various store characteristics influence customer behavior and sales outcomes. This consolidated dataset is vital for conducting in-depth explorations, trend analyses, and data-driven decision-making.

###Cleaning ***merged_dataset***

```python
# Convert the 'date' column to a datetime type
merged_data['date'] = pd.to_datetime(merged_data['date'])
```

### Sample the ***merged_data*** dataframe

```python
merged_data
```

###Grouping ***by_month***

```python
# Group data by month
by_month = merged_data.groupby(merged_data['date'].dt.to_period('M'))['transactions'].sum().reset_index()
# Convert 'date' to string representation of month
by_month['date'] = by_month['date'].dt.strftime('%Y-%m')
by_month = by_month.drop(by_month.index[-1])
```

###Display ***by_month***

```python
df = by_month
fig = px.line(df, x='date', y='transactions')
# Compute the cumulative mean of the 'transactions' column
by_month['cumulative_mean'] = by_month['transactions'].cumsum() / (by_month.index + 1)

# Use Plotly to create the line chart for transactions and cumulative mean
fig = px.line(by_month, x='date', y='transactions', labels={'date': 'Month', 'transactions': 'Total Transactions'},
              title='Total Transactions per Month')

fig.add_scatter(x=by_month['date'], y=by_month['cumulative_mean'], mode='lines', name='Cumulative Mean (Transactions)')

fig.show()
```

#Machine **Learning** (Time Series Data)

##Importing **Prophet**

```python
from prophet import Prophet
```

Prophet is a procedure for forecasting time series data based on an additive model where non-linear trends are fit with yearly, weekly, and daily seasonality, plus holiday effects.

## Data / Week

### Grouping **by_week**

```python
# Group data by week
by_week = merged_data.groupby(merged_data['date'].dt.to_period('W'))['transactions'].sum().reset_index()

# Convert 'date' to string representation of the week's start date
by_week['date'] = by_week['date'].apply(lambda x: x.start_time.strftime('%Y-%m-%d'))

by_week = by_week.drop(by_week.index[-1])
```

### Display **by_week**

```python
df = by_week
fig = px.line(df, x='date', y='transactions')
# Compute the cumulative mean of the 'transactions' column
by_week['cumulative_mean'] = by_week['transactions'].cumsum() / (by_week.index + 1)

# Use Plotly to create the line chart for transactions and cumulative mean
fig = px.line(by_week, x='date', y='transactions', labels={'date': 'Weeks', 'transactions': 'Total Transactions'},
              title='Total Transactions per Week')

fig.add_scatter(x=by_week['date'], y=by_week['cumulative_mean'], mode='lines', name='Cumulative Mean (Transactions)')

fig.show()
```

### Prepare Data for Model Fitting

```python
df = by_week[['date', 'transactions']]
df.columns = ['ds', 'y']
```

```python
m = Prophet()
```

```python
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=500)
```

```python
forecast = m.predict(future)
```

```python
forecast.describe()
```

```python
plot = m.plot(forecast)
```
